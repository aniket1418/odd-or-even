```python
import tensorflow as tf
import numpy as np
```

```python
x_nums = np.random.randint(low=0, high=2**31 + 1, size=(12000, 1)) #random numbers from 0 to 2^31
y = np.uint8(x_nums % 2 == 1) #1 if odd, 0 if even
x = np.unpackbits(x_nums.astype(">i4").view(np.uint8), axis=1) #convert numbers to their 32 bit binary representation

#split training and test sets
x_train, x_test = x[:10000], x[10000:]
y_train, y_test = y[:10000], y[10000:]
```

```python
#define placeholders for inputs and labels
X = tf.placeholder(shape=(None, 32), dtype=tf.float32)
Y = tf.placeholder(shape=(None, 1), dtype=tf.float32)

#Create neural network with a single hidden layer which has 4 units
W1 = tf.Variable(np.random.randn(32, 4), dtype=tf.float32)
b1 = tf.Variable(np.random.randn(1, 4), dtype=tf.float32)

#Single output neuron
W2 = tf.Variable(np.random.randn(4, 1), dtype=tf.float32)
b2 = tf.Variable(np.random.randn(1, 1), dtype=tf.float32)


#define the operations that occur between the layers
Z1 = tf.add(tf.matmul(X, W1), b1)
A1 = tf.nn.relu(Z1)
Z2 = tf.add(tf.matmul(A1, W2), b2)

#apply sigmoid function on NN output and minimize cross entropy loss
cost = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(labels=Y, logits=Z2))
optimizer = tf.train.AdamOptimizer(learning_rate=0.01).minimize(cost)
```

```python
epochs = 5
mb_size = 50
sess = tf.Session()
sess.run(tf.global_variables_initializer())

for epoch in range(epochs):
    #shuffle the data and labels every epoch
    idx = np.random.permutation(x.shape[0])
    x, y = x[idx], y[idx]
    
    print("Epoch " + str(epoch + 1) + ":")
    for i in range(int(x_train.shape[0] / mb_size)):
        _x = x_train[i*mb_size:(i+1)*mb_size, :]
        _y = y_train[i*mb_size:(i+1)*mb_size, :]
        print("Minibatch " + str(i + 1) + " Cost", sess.run(cost, {X: _x, Y:_y}))
        sess.run(optimizer, {X: _x, Y:_y})
    print()
print("Test Cost", sess.run(cost, {X: x_test, Y:y_test}))
sess.close()
```
